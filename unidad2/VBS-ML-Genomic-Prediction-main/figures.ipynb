```r
# First Figure 3 so load the results of each model first
naiveml_split1 <- read.table("Naive-ML/result_split1/predictions.txt", sep="\t", header=TRUE)
naiveml_split3 <- read.table("Naive-ML/result_split3/predictions.txt", sep="\t", header=TRUE)

bayesA_split1 <- read.table("res_BA/split1/2014_4_BA.csv", sep=",", header=TRUE)
bayesA_split2 <- read.table("res_BA/split2/2014_4_BA.csv", sep=",", header=TRUE)
bayesA_split3 <- read.table("res_BA/split3/2014_4_BA.csv", sep=",", header=TRUE)
bayesA_split4 <- read.table("res_BA/split4/2014_4_BA.csv", sep=",", header=TRUE)
bayesA_split5 <- read.table("res_BA/split5/2014_4_BA.csv", sep=",", header=TRUE)

bayesB_split1 <- read.table("res_BB/split1/2014_2_BB.csv", sep=",", header=TRUE)
bayesB_split2 <- read.table("res_BB/split2/2014_2_BB.csv", sep=",", header=TRUE)
bayesB_split3 <- read.table("res_BB/split3/2014_2_BB.csv", sep=",", header=TRUE)
bayesB_split4 <- read.table("res_BB/split4/2014_2_BB.csv", sep=",", header=TRUE)
bayesB_split5 <- read.table("res_BB/split5/2014_2_BB.csv", sep=",", header=TRUE)

vbs_ml <- read.table("VBS-ML/run_output/predictions.txt", sep="\t", header=TRUE)
```
